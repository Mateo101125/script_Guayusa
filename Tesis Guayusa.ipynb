```python
f=data.Identificador.str.contains("Flux", case=False, regex=True)
data_flux= data[f]
# draw pointplot
sns.pointplot(x= "Sitio", 
              y = "Intensidad",
              hue = "Luz",
              data = data_flux)
# show the plot
plt.show()
# This code is contributed 
# by Deepanshu Rustagi.
```

```python
lf=data_flux.Luz.str.contains("ligth", case=False, regex=True)
data_ligth_f= data_flux[lf]
sf=data_flux.Luz.str.contains("shadow", case=False, regex=True)
data_shadow_f= data_flux[sf]
```

```python
min_lf= data_ligth_f.min()
max_lf=data_ligth_f.max()
print (max_lf)
print (min_lf)
```

```python
min_sf= data_shadow_f.min()
max_sf=data_shadow_f.max()
print (max_sf)
print (min_sf)
```

```python
p1=data_flux.Sitio.str.contains("C", case=False, regex=True)
p2=data_flux.Luz.str.contains("ligth", case=False, regex=True)
data_p1= data_flux[p1 & p2]
print(data_p1)
mean_ligth= data_p1.mean(numeric_only= True)
max_ligth= data_p1.max()
min_ligth= data_p1.min()
print(mean_ligth)
print(max_ligth)
print(min_ligth)
```
